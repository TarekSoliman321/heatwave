```python
y = nc.rlat #reading lattitudes
x = nc.rlon #reading longitude
#select variables at timestep
im_data = nc.isel(time=0,rlon=131, rlat=83)
temp = im_data['tasmax'].variable.values
print("This is the Temperature (K) in Egypt:", temp)
```

Ectracting data in a dictionary format

```python
dic = {}
for i in range(1825):
  im_data = nc.isel(time=i,rlon=131, rlat=83)
  day = i
  temp = im_data['tasmax'].variable.values
  dic[i] = temp-273.15 
  # for values in dic.values():
  #  print(values)

print (dic)
```

Converting dictionary to CSV file

```python
import csv
with open('TempMax.csv', 'w', newline='') as f:
    fieldnames = ['Day','Temp']
    writer = csv.writer(f)
    for row in dic.items():
        writer.writerow(row)
```
